```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as pe
```

```python
tips=sns.load_dataset('tips')
gap=pe.data.gapminder()
titanic=sns.load_dataset('titanic')
```

# Relational Plot

fig. level plot --> relplot()

axis level plot --> scatterplot() , lineplot()

# 1.              Scatter Plot

Scatter Plot (Axis Level)

```python
sns.scatterplot(tips,x='total_bill',y='tip')
```

hue parameter

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex')
```

style parameter

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex',style='time')
```

size parameter

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',col='sex',row='day')
```

# 2. KDE plot :
Rather than discrete bins , a KDE Plot smooth the observation with a Gaussian Kernel, Producing a continous density estimate

```python
sns.kdeplot(tips,x='total_bill')
```

```python
sns.displot(tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

# 3. Rug Plot :-
Plot marginal distribution by drawing ticks along the  and y axis

This Function is intended to complement of other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.kdeplot(tips,x='total_bill')
sns.rugplot(tips,x='total_bill')
```

# Bivariate Histogram :- 
A bivariate histogram bins the data within rectangles that tile the plot

and then show the count of observations within each rectangles with the fill color

```python
sns.histplot(tips,x='total_bill',y='tip')
```

# Bivariate KDE plot :-
A bivariate KDE plot smoothes the (x,y) observations with a 2D Gaussian

```python
sns.kdeplot(tips,x='total_bill',y='tip')
```

## Matrix Plot

# 1. Heatmap :-
Plot rectangular data as a color-encoded matrix

It is Axis level plot

There is no figure level in heatmap

```python
gap
```

Converting Data From long to wide data format

```python
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

annot , linewidth , cmap

```python
temp=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp,annot=True,linewidths=0.4,cmap='summer')
```

# 2. Clustermap :-
Plot a Matrix dataset as a hierarchically-clustered heatmap.

Tells the Simillarity between 2 or more columns

This function requires scipy to be available

```python
iris=pe.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
